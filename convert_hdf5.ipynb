# Converting the data

```python
import h5py
import pandas as pd
import numpy as np

# open the HDF5 file
hdf5_file_path = 'DC282_26761_0_test2.h5'

csv_file_path = 'output' + hdf5_file_path[:-3] + '.csv'

data_records = []

# function to extract data from the datasets
def extract_data_from_group(group_name, group):
    '''Extract and flatten the data from a group's keys '''
    for key in group.keys():
        dataset = group[key]
        if isinstance(dataset, h5py.Dataset):
            # flatten the dataset if it has more than 1 dimension
            flat_data = dataset[()].flatten()
            data_records.append({
                'Group': group_name,
                'Dataset': key,
                'Data': flat_data
            })

# iterate through each acquisition group and extract their data
with h5py.File(hdf5_file_path, 'r') as f:
    for acquisition_name in f.keys():
        acquisition_group = f[acquisition_name]
        if isinstance(acquisition_group, h5py.Group):
            extract_data_from_group(acquisition_name, acquisition_group)

# convert the records to a pandas dataframe
df = pd.DataFrame(data_records)

# expand the 'Data' field into separate columns
data_expanded = pd.DataFrame(df['Data'].tolist())
df = pd.concat([df.drop('Data', axis=1), data_expanded], axis=1)

# save to csv
df.to_csv(csv_file_path, index=False)

print(f"Data successfully written to {csv_file_path}.")
```
